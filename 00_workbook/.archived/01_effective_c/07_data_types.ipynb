# Notes

So C has the following types:

1. bool
2. char
3. unsigned char
4. signed char
5. wchar_t
6. short int
7. int
8. long int
10. long long int
11. unsigned short int
12. unsigned int
13. unsigned long int
14. unsigned long long int
15. enum
16. float
17. double
18. long double
19. void
20. function types
21. derived types (pointers, arrays, type definitions, structs, unions).
    - Pointer type - this is derviced from the function or object type that it points to, called the referenced type. (`*` is called the indirection operator and its use to dereference a pointer, `&` is called the address-of operator and is used to reference a pointer.
    - Array type - a contiguously allocated sequence of objects that have the same designated element type.
    - Type definitions - these are aliases for other types.
    - Structures - These contain sequentially allocated memory objects of different types.
    - Union types - These define a series of types, but the object can only contain one type at a time.

There are more types in special header files. like `stdint.h`, `inttypes.h`, `stdbool.h`. some limits are specified in `limits.h`.

In C, a function with an empty parameter list means it accepts ANY number of arguments of ANY type. So you need to specifically give it a `void` parameter to denote that it accepts no parameters.

## typedefs

You can use typedefs to declare aliases for types.

You can do this three ways:
1. typedef for a variable type alias
    - typedef signed char schar_t;

2. typedef for a pointer type alias
    - typedef signed char *schar_p;
  
3. typedef for a function type alias
    - typedef signed char (*schar_fd)(void);

## structures

The following declares a struct type with an alias sigline, and a pointer alias `sigline_p`.
```c
struct {
    int signum;
    char signame[20];
    char sigdesc[100];
} sigline, *sigline_p;
```

## Tags

Tags are a special way of declaring structs and union types.

```c
struct Person {
    int age;
    char name[100] name;
}
```

In the above, `Person` is a tag. We can declare a `Person` object as follows:

```c
struct Person lusayo;
lusayo.age = 27;
lusayo.name = "Lusayo";
```

You can combine typedefs and structs to make essentially named types:

```c
typedef struct {
    int age;
    int name;
} Person;

Person angelise;

angelise.age = 11;
angelise.name = "Angelise";
```

## Type Qualifiers

1. const - a qualifier that restricts redeclarations.
2. volatile - a qualifier for locations in memory that may be modified without the compiler's knowledge. These are memory locations that map to input/output ports and input ports like the system clock.
3. restrict - this is used to promote optimization.
